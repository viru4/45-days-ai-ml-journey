```python
import numpy as np
import random

# Grid settings
grid_size = 4
actions = ["up", "down", "left", "right"]

# Initialize Q-table
Q = {}
for i in range(grid_size):
    for j in range(grid_size):
        Q[(i, j)] = {a: 0 for a in actions}

# Rewards
goal = (3, 3)
obstacle = (1, 1)

def get_next_state(state, action):
    i, j = state

    if action == "up":
        i = max(i - 1, 0)
    elif action == "down":
        i = min(i + 1, grid_size - 1)
    elif action == "left":
        j = max(j - 1, 0)
    elif action == "right":
        j = min(j + 1, grid_size - 1)

    return (i, j)

def get_reward(state):
    if state == goal:
        return 10
    elif state == obstacle:
        return -10
    else:
        return -1  # small penalty to encourage faster path

# Hyperparameters
alpha = 0.1
gamma = 0.9
epsilon = 1.0
decay = 0.995

# Training
for episode in range(200):

    state = (0, 0)

    while state != goal:

        # ε-greedy
        if random.random() < epsilon:
            action = random.choice(actions)
        else:
            action = max(Q[state], key=Q[state].get)

        next_state = get_next_state(state, action)
        reward = get_reward(next_state)

        # Q update
        max_next_q = max(Q[next_state].values())
        Q[state][action] += alpha * (reward + gamma * max_next_q - Q[state][action])

        state = next_state

    epsilon = max(0.1, epsilon * decay)

print("Learned Q-values:")
for state in Q:
    print(state, Q[state])
```

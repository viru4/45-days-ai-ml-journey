```python
def update_q(Q, state, action, reward, next_state, alpha, gamma):
    """
    Q-learning update rule implementation
    """

    # Get current Q value
    current_q = Q[state][action]

    # Get max Q value for next state
    max_next_q = max(Q[next_state].values())

    # Q-learning formula
    new_q = current_q + alpha * (reward + gamma * max_next_q - current_q)

    # Update Q-table
    Q[state][action] = new_q

    return Q


# Example usage
Q = {
    "s1": {"left": 2, "right": 5},
    "s2": {"left": 3, "right": 8}
}

alpha = 0.1
gamma = 0.9

Q = update_q(Q, "s1", "left", reward=2, next_state="s2", alpha=alpha, gamma=gamma)

print(Q)
```

```python
import random

actions = ["left", "right"]

# Example Q-table
Q = {
    "s1": {"left": 2, "right": 5},
    "s2": {"left": 3, "right": 8}
}

alpha = 0.1
gamma = 0.9

# Q-Learning
def q_learning(Q):
    state = "s1"
    action = "left"
    reward = 2
    next_state = "s2"

    max_next_q = max(Q[next_state].values())

    Q[state][action] += alpha * (reward + gamma * max_next_q - Q[state][action])
    return Q

# SARSA
def sarsa(Q):
    state = "s1"
    action = "left"
    reward = 2
    next_state = "s2"

    # choose next action (ε-greedy simulation)
    next_action = random.choice(actions)

    Q[state][action] += alpha * (reward + gamma * Q[next_state][next_action] - Q[state][action])
    return Q

print("Q-Learning:", q_learning(Q.copy()))
print("SARSA:", sarsa(Q.copy()))
```

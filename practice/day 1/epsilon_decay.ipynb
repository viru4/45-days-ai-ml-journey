```python
import random

epsilon = 1.0
decay = 0.9
min_epsilon = 0.1

Q = {"left": 2, "right": 10}
actions = list(Q.keys())

for episode in range(10):
    
    if random.random() < epsilon:
        action = random.choice(actions)
    else:
        action = max(Q, key=Q.get)

    print(f"Episode {episode}, Epsilon: {epsilon:.2f}, Action: {action}")
    
    # decay epsilon
    epsilon = max(min_epsilon, epsilon * decay)
```

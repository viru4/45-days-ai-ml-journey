```python
# tabular_q_learning_demo.py

import numpy as np
import random

# Define environment
n_states = 5   # small state space
n_actions = 2  # left, right

# Initialize Q-table
Q = np.zeros((n_states, n_actions))

# Hyperparameters
alpha = 0.1      # learning rate
gamma = 0.9      # discount factor
epsilon = 0.3    # exploration

# Simple environment transition
def step(state, action):
    """
    Simulates environment dynamics
    """
    if action == 0:  # left
        next_state = max(0, state - 1)
    else:  # right
        next_state = min(n_states - 1, state + 1)

    reward = 1 if next_state == n_states - 1 else 0
    return next_state, reward

# Training loop
episodes = 50

for episode in range(episodes):
    state = random.randint(0, n_states - 1)

    for _ in range(10):
        
        # ε-greedy action selection
        if random.random() < epsilon:
            action = random.choice([0, 1])
        else:
            action = np.argmax(Q[state])

        next_state, reward = step(state, action)

        # Q-learning update
        best_next_action = np.argmax(Q[next_state])

        Q[state][action] = Q[state][action] + alpha * (
            reward + gamma * Q[next_state][best_next_action] - Q[state][action]
        )

        state = next_state

# Print learned Q-table
print("Q-table:")
print(Q)
```

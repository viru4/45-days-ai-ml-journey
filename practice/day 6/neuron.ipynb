```python
import numpy as np

# Step 1: Define activation functions

def sigmoid(x):
    """
    Sigmoid activation function
    Output range: (0, 1)
    """
    return 1 / (1 + np.exp(-x))


def relu(x):
    """
    ReLU activation function
    Output: max(0, x)
    """
    return np.maximum(0, x)


# Step 2: Define a Neuron class

class Neuron:
    def __init__(self, input_size):
        """
        Initialize weights and bias
        
        input_size: number of input features
        """
        self.weights = np.random.randn(input_size)  # random weights
        self.bias = np.random.randn()               # random bias

    def forward(self, inputs, activation="sigmoid"):
        """
        Perform forward pass
        
        inputs: list or numpy array
        activation: type of activation function
        """
        # Step 1: Weighted sum
        z = np.dot(self.weights, inputs) + self.bias

        # Step 2: Apply activation
        if activation == "sigmoid":
            return sigmoid(z)
        elif activation == "relu":
            return relu(z)
        else:
            return z  # linear (no activation)


# Step 3: Test the neuron

if __name__ == "__main__":
    
    # Example input (3 features)
    inputs = np.array([0.5, 0.2, 0.1])
    
    # Create neuron
    neuron = Neuron(input_size=3)
    
    # Forward pass
    output = neuron.forward(inputs, activation="sigmoid")
    
    print("Weights:", neuron.weights)
    print("Bias:", neuron.bias)
    print("Output:", output)
```
